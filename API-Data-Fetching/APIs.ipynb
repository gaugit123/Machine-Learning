### Data Fetching through API

```python
import pandas as pd
import requests
```

```python
url = "https://ai-movie-recommender.p.rapidapi.com/api/trending" 

headers = {
	"x-rapidapi-key": "Your API Key",
    "x-rapidapi-host": "ai-movie-recommender.p.rapidapi.com",
}
response = requests.get(url, headers=headers)
data = response.json()['results']
# print(data)
# print(type(data)) #always check the type of data you get from API, since it is dictionary you need to convert it into dataframe, 

# if it directly returns list then use below code
# df = pd.DataFrame(data)

# else
df = pd.DataFrame(data)
```

```python
new_df = df[['id','title','overview','popularity','release_date','vote_average','vote_count']]  
```

```python
# df[df['original_language'] ==   'en']
new_df.head() #now this is the dataframe of only one page, you can loop through all the pages and get the data of all the movies. Total there is 500 pages, you can use for loop to get the data of all the pages and then concatenate all the dataframes to get the final dataframe of all the movies.
```

```python
final_data = pd.DataFrame() #empty dataframe to store the data of all the pages
```

```python
all_pages = [] #list to store the data of all the pages
for page in range(1, 4):# took only 3 pages to avoid hitting the API limit.
    response = requests.get(url, headers=headers, params={'page': page})
    df = pd.DataFrame(response.json()['results'])[['id','title','overview','popularity','release_date','vote_average','vote_count']]
    all_pages.append(df)
final_data = pd.concat(all_pages, ignore_index=True)
```
