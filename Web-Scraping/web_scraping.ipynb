```python
len(soup.find_all('article', class_='product_pod')) #this will return a list of all the articles with class product_pod, you can loop through this list and get the data of each book. since len is 20 it means there are 20 books in each page, you can loop through all the pages and get the data of all the books. Total there is 50 pages, you can use for loop to get the data of all the pages and then concatenate all the dataframes to get the final dataframe of all the books.
```

```python
titles = []
ratings = []
prices = []
```

```python
#for final title, rating and price, you can loop through all the pages and get the data of all the books and then concatenate all the dataframes to get the final dataframe of all the books.
for j in range(1, 5):
     headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
     url = f'https://books.toscrape.com/catalogue/page-{j}.html'
     webpage = requests.get(url,headers=headers)
     soup = BeautifulSoup(webpage.content, 'html.parser')

     #for title
     for i in soup.find_all('h3'):
        titles.append(i.a['title'].strip())

     #for rating
     for i in soup.find_all('p', class_='star-rating'):
        ratings.append(i['class'][1])
        
     #for price
     for i in soup.find_all('p', class_='price_color'):
        prices.append(i.text) 
```

```python
len(titles), len(ratings), len(prices) #since there are 20 books in each page and we have scraped 4 pages, the length of each list should be 80. if there is any error in getting the title, rating or price of any book, it will be stored as None in the list. you can check the length of each list to see if there is any error in getting the data of any book.
```

```python
#for genere we need to go to the book page and get the genre of the book, since the genre is not available in the main page, we need to go to the book page and get the genre of the book, we can use the same headers and make a request to the book page and then parse the html content of the book page and get the genre of the book.
```

```python
soup2 = BeautifulSoup(webpage2.content, 'html.parser')
```

```python
soup2.find('ul', class_='breadcrumb').find_all('li')[2].text.strip()
 #this will return the genre of the book, since the genre is in the breadcrumb and it is the third li tag, we can get it by using find_all('li')[2] since the index starts from 0.
```

```python
#for description
soup2.find('div', id='product_description').find_next_sibling('p').text.strip() #this will return the description of the book, since the description is in the p tag which is the next sibling of the div tag with id product_description, we can get it by using find_next_sibling('p').strip() is used to remove the extra spaces from the description.BeautifulSoup searches inside the div if we write find('p') instead of find_next_sibling('p') it will return None since there is no p tag inside the div tag with id product_description.        
```

```python
genre = []
description = []
```

```python
## final for genre and description,
for j in range(0, 80):
    book_num = 1000 - j
    book = titles[j].replace(':', '').replace(',', '').replace(' ', '-').lower()
 #since the book name is in the url and it is in lowercase and spaces are replaced with hyphen, we need to replace the spaces with hyphen and convert the book name to lowercase.
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    
    url= 'https://books.toscrape.com/catalogue/{}_{}/index.html'.format(book,book_num)

    webpage2 = requests.get(url,headers=headers).text
    soup2 = BeautifulSoup(webpage2, 'html.parser')
    
    breadcrumb = soup2.find('ul', class_='breadcrumb')
    description_div = soup2.find('div', id='product_description')

    if breadcrumb:
        items = breadcrumb.find_all('li')
        if len(items) > 2:
            genre.append(items[2].text.strip())
        else:
            genre.append(None)
    else:
        genre.append(None)

    if description_div:
        description.append(description_div.find_next_sibling('p').text.strip())
    else:
        description.append(None)
```

```python
print(len(genre), len(description))
```

```python
#final dataframe
final_df = pd.DataFrame({'Title': titles, 'Rating': ratings, 'Price': prices, 'Genre': genre, 'Description': description})#this will create a dataframe with the title, rating, price, genre and description of the books, since we have 1000 (here only 80) books in the website, the length of each list should be 1000 (here only 80), if there is any error in getting the genre or description of any book, it will be stored as None in the dataframe.
```
